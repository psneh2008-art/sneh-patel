```python
global_summary = {}
def show_basic_stats(data):
    """Show basic statistics using built-in functions"""
    print("Length:", len(data))
    print("Sum:", sum(data))
    print("Min:", min(data))
    print("Max:", max(data))

def calculate_average(data):
    """Calculate average of dataset"""
    return sum(data) / len(data) if len(data) > 0 else 0


def find_duplicates(data):
    """Find duplicate values in dataset"""
    duplicates = []
    for i in data:
        if data.count(i) > 1 and i not in duplicates:
            duplicates.append(i)
    return duplicates


def unique_values(data):
    """Find unique values"""
    unique = []
    for i in data:
        if i not in unique:
            unique.append(i)
    return unique


# ==========================
# *args, **kwargs, __doc__
# ==========================
def display_args(*args):
    """Display all arguments passed"""
    print("Values:", args)


def display_kwargs(**kwargs):
    """Display dataset characteristics"""
    for key, value in kwargs.items():
        print(key, ":", value)

def factorial(n):
    """Recursive factorial"""
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)


def fibonacci(n):
    """Recursive Fibonacci"""
    if n <= 1:
        return n
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

def filter_threshold(data, threshold):
    """Filter values greater than threshold"""
    filtered = list(filter(lambda x: x > threshold, data))
    return filtered


def map_square(data):
    """Square each value using lambda"""
    squared = list(map(lambda x: x * x, data))
    return squared
def update_summary(data):
    """Update global summary"""
    global global_summary
    global_summary = {
        "count": len(data),
        "mean": calculate_average(data)
    }


def dataset_statistics(data):
    """Return min, max, average"""
    return min(data), max(data), calculate_average(data)

def sort_list(data):
    """Sort list using sort()"""
    data.sort()
    return data

def sorted_2d_list(data):
    """Sort 2D list using sorted()"""
    return sorted(data, key=lambda x: x[0])
def display_2d(data):
    """Display 2D list in grid format"""
    for row in data:
        print(" ".join(str(x) for x in row))
def main():
    print("==== Functional Treat Project ====")
    data1D = [5, 2, 8, 2, 9, 1]
    data2D = [[3, 2, 1], [9, 5, 6], [4, 7, 8]]

    while True:
        print("\nMain Menu")
        print("1. Show Basic Stats")
        print("2. User Defined Functions")
        print("3. *args and **kwargs")
        print("4. Recursion")
        print("5. Lambda Functions")
        print("6. Global Keyword")
        print("7. Return Multiple Values")
        print("8. Sorting")
        print("9. Display 2D Array")
        print("0. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            show_basic_stats(data1D)

        elif choice == "2":
            print("Average:", calculate_average(data1D))
            print("Duplicates:", find_duplicates(data1D))
            print("Unique:", unique_values(data1D))

        elif choice == "3":
            display_args(1, 2, 3, 4, 5)
            display_kwargs(length=len(data1D), mean=calculate_average(data1D))
            print(display_args.__doc__)

        elif choice == "4":
            n = 5
            print("Factorial of", n, "=", factorial(n))
            print("Fibonacci of", n, "=", fibonacci(n))

        elif choice == "5":
            print("Filter > 4:", filter_threshold(data1D, 4))
            print("Square each:", map_square(data1D))

        elif choice == "6":
            update_summary(data1D)
            print("Global Summary:", global_summary)

        elif choice == "7":
            mn, mx, avg = dataset_statistics(data1D)
            print("Min:", mn, "Max:", mx, "Avg:", avg)

        elif choice == "8":
            print("Original:", data1D)
            print("Sorted (in-place):", sort_list(data1D))
            print("Sorted 2D:", sorted_2d_list(data2D))

        elif choice == "9":
            display_2d(data2D)

        elif choice == "0":
            print("Exiting program...")
            break

        else:
            print("Invalid choice, try again!")
main()
```
